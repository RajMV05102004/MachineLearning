The genre is in a dictionary and the dictionary itself is stored as String. So, to deal with this, we are going to create a function

```python
import ast
#Abstract Syntax Tree
```

```python
movies['genres'].iloc[0]
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(obj):
    L=[]
    j=0
    for i in ast.literal_eval(obj):
        if(j<3):
            L.append(i['name'])
            j=j+1
    return L 
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            L.append(i['name'])
            break
    return L
```

```python
def process_movie_name(movie_name):
    # Split the movie name into words
    words = movie_name.split()
    L=[]
    main_name=movie_name.split(':')
    
    # Remove unwanted characters (':', '-', ' ') from the combined name
    first_name = ''.join(words[0]).replace(':', '').replace('-', '').replace(' ', '')
    main_name_2=''.join(main_name[0].replace(' ',''))
    #print(main_name_2)
    L.append(first_name)
    if(main_name_2 != first_name):
        L.append(main_name_2)
    return [item for item in L if item != '']

# Example usage
movie_name = "Avengers: Age of Ultron"
result = process_movie_name(movie_name)
print(result)

movie_name2 = "Pirates of the Caribbean: At World's End"
result2 = process_movie_name(movie_name2)
print(result2)
```

```python
def remove_empty_strings(input_list):
    return [item for item in input_list if item != '']
```

```python
df['tags']=df['tags'].apply(lambda x: " ".join(x))
```

```python
df.sample(5)
```

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps=  PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
for i in cv.get_feature_names_out():
    print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity[1]
```

```python
similarity[0]
```

```python
def recommend(movie):
    index = df[df['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key= lambda x:x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
